# Datos sobre el covid-19 en Colombia

```python
import psycopg2
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
conn = psycopg2.connect("postgresql://admin:secret@postgres:5432/postgres")
```

```python
SQL_Query = pd.read_sql_query(
'''SELECT edad, "Ubicación del caso", estado,
    "Nombre departamento", "Fecha de inicio de síntomas",
    "Fecha de muerte", "Fecha de recuperación", sexo, recuperado FROM covid_data''', conn)
```

```python
df = pd.DataFrame(SQL_Query, columns=['edad',
                                      'Ubicación del caso',
                                      'estado',
                                      'Nombre departamento',
                                      'Fecha de inicio de síntomas',
                                      'Fecha de muerte',
                                      'Fecha de recuperación',
                                      'sexo',
                                      'recuperado']).dropna()
```

```python
sns.catplot(x="edad", y="estado", kind="box", data=df)
```

```python
sns.countplot(y="sexo", data=df)
```

```python
sns.displot(df["recuperado"])
```
